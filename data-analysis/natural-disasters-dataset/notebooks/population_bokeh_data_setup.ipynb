```python
import pandas as pd
from pathlib import Path
```

```python
em_dat_iso_path = Path("../Resources/disaster_data/emdat_cleaned_iso.json")

emdat_df = pd.read_json(em_dat_iso_path)
```

```python
print(emdat_df.shape)

emdat_df.head()
```

```python
country_population_path = Path("../Resources/Country_Population_Data/country-by-population.json")

country_pop_df = pd.read_json(country_population_path)
```

```python
#population is people as of 2020 world bank data according to git repo history
print(country_pop_df.shape)
country_pop_df.head()
```

```python
country_population_density_path = Path("../Resources/Country_Population_Data/country-by-population-density.json")

country_pop_density_df = pd.read_json(country_population_density_path)
```

```python
#density is people per square kilmoeter as of 2020 world bank data according to git repo history
country_pop_density_df.head()
```

```python
country_surface_area_path = Path("../Resources/Country_Population_Data/country-by-surface-area.json")

country_surface_area_df = pd.read_json(country_surface_area_path)
```

```python
#in square kilometers
country_surface_area_df.head()
```

```python
country_df = country_pop_df.merge(country_pop_density_df, on='country')
print(country_df.shape)

country_df = country_df.merge(country_surface_area_df, on='country')

print(country_df.shape)
country_df.head()
```

```python
#get iso numeric country code and merge with country with df

country_iso_num_path = Path("../Resources/Country_Population_Data/country-by-iso-numeric.json")

country_iso_num_df = pd.read_json(country_iso_num_path)
country_iso_num_df.head()
```

```python
country_df = country_df.merge(country_iso_num_df, on='country')
print(country_df.shape)
country_df.head(15)
```

```python
country_df[country_df['country']=='Russia']
```

```python
#for some reason, russia didn't have a valid iso, assigned the correct one
country_df.at[174, 'iso'] = 643
```

```python
country_df[country_df['country']=='Russia']
```

```python
#get slim 3 data to df and merge on iso numeric 

country_slim_path = Path("../Resources/Country_Population_Data/slim-3.json")

country_slim_df = pd.read_json(country_slim_path)
country_slim_df.head(15)
```

```python
country_slim_df[country_slim_df['country-code']==643]
```

```python
country_df = country_df.merge(country_slim_df, left_on='iso', right_on='country-code')
print(country_df.shape)
country_df.head()
```

```python
country_df[country_df['country-code']==643]
```

```python
country_df = country_df[['alpha-3', 'population', 'density', 'area']]
country_df.head()
```

```python
#original merge on 'country' gave us only 57991 rows, which lost us about 20k
# emdat_country_df = emdat_df.merge(country_df, how='inner', left_on='Country', right_on='country')


emdat_country_df = emdat_df.merge(country_df, how='inner', left_on='ISO', right_on='alpha-3')

print(emdat_country_df.shape)

emdat_country_df.head()
```

```python
#remove duplicate ISO column, and can potentially remove more later
clean_df = emdat_country_df[['Disaster #', 'Subgroup', 'Type', 'Subtype', 'Country', 'ISO', 'Region', 'Geolocation', 'Magnitude', 'Lat', 'Lng',	'Year',	'Month', 'Deaths', 'Injuries', 'Total Affected', 'Insured Damage (Adjusted)', 'Total Damage (Adjusted)', 'population', 'density', 'area']]
```

```python
clean_df.to_csv('../Resources/RileyAnalysis/emdat_population.csv')
clean_df.to_json('../Resources/RileyAnalysis/emdat_population.json', orient='records', indent=4)
```
