```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf
```

```python
# Import our input dataset
file_path = r'C:\Users\micha\Desktop\Data_Analytics\UTA-BootCamp-Final-Project\Resources\matches_cleaned.csv'
matches_cleaned = pd.read_csv(file_path)
matches_cleaned
```

```python
matches_cleaned.columns
```

```python
#drop the columns that we don't need
drop_cols = ['Unnamed: 0',
             'tourney_id',
             'tourney_name',
             'draw_size',
             'tourney_date',
             'match_num',
             'player_id',
             'player_seed',
             'player_name',
             'player_ht',
             'score',
             'best_of',
             'set_3_score',
             'set_4_score',
             'set_5_score']

matches_cleaned.drop(columns=drop_cols, inplace=True)


#check the remaining columns
matches_cleaned.columns
```

```python
#drop missing values
matches_cleaned = matches_cleaned.dropna()
```

```python
#view the new dataframe
matches_cleaned
```

```python
#view the missing values by column to make sure there are NONE
print(matches_cleaned.isna().sum())
```

```python
#create a list of variables to categorically encode
cols_to_encode = ['surface',
               'tourney_level',
               'player_hand',
               'player_ioc',
               'round']
```

```python
#check the number of unique values in the categorical columns
#this will determine how many new columns are generated per column in 'cols_to_encode'
matches_cleaned[cols_to_encode].nunique()
```

```python
#create a dataframe that isolates these columns
matches_cleaned_cols_to_encode = matches_cleaned[cols_to_encode]
```

```python
#view the dataframe
matches_cleaned_cols_to_encode
```

```python
#check the columns
matches_cleaned.columns
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(matches_cleaned_cols_to_encode))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names_out(cols_to_encode)
encode_df
```

```python
#check the columns
encode_df.columns
```

```python
# Merge one-hot encoded features and drop the originals
matches_cleaned.drop(cols_to_encode, axis=1, inplace=True)
matches_cleaned_2 = matches_cleaned.reset_index().merge(encode_df, left_index=True, right_index=True)

matches_cleaned_2
```

```python
# Split our preprocessed data into our features and target arrays
y = matches_cleaned_2["winner"].values

X = matches_cleaned_2.drop(columns='winner')

##X = matches_cleaned.drop(columns=["winner", "Unnamed", "tourney_id", "tourney_name", ""]).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
print(matches_cleaned_2.isna().sum())
```

```python
X_train.columns
```

```python
# Support vector machine linear classifier
from sklearn.svm import SVC
model = SVC(kernel='linear')
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
predictions
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=["loser", "winner"]))
```

```python
y_test
```

```python
# evaluate model
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, accuracy_score, classification_report

#proba_predictions = model.predict_proba(X_test)[:, 1]

accuracy = accuracy_score(y_test, predictions)
precision = precision_score(y_test, predictions)
recall = recall_score(y_test, predictions)
f1 = f1_score(y_test, predictions)
#roc_auc = roc_auc_score(y_test, proba_predictions)
conf_matrix = confusion_matrix(y_test, predictions)

print(f'Accuracy: {accuracy:.2f}')
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1-Score: {f1:.2f}')
#print(f'ROC AUC: {roc_auc:.2f}')
print(f'Confusion Matrix:\n{conf_matrix}')
print(f'\nClassification Report:\n{classification_report(y_test, predictions)}')

```
